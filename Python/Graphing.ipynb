```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

a = int(input("Enter the value of the amplitude. "))
b = int(input("Enter the value of the period coeffcient. "))
c = int(input("Enter the value of the vertical shift. "))

x = np.arange(-np.pi,np.pi,0.1)
axes = plt.gca()
axes.set_ylim([-a - c,a + c])

y1 = a * np.sin(b * x) + c 
y2 = a * np.cos(b * x) + c
y3 = a * np.tan(b * x) + c 
plt.plot(x,y1,x,y2,x,y3,'.');
```
